```python
import matplotlib.pyplot as plt

# Adjust the size of the entire figure
plt.figure(figsize=(12, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(df["country"], df["beer_servings"], "r-")
plt.xlabel("Country")
plt.ylabel("beer_servings")

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(df["continent"], df["wine_servings"], "y")
plt.xlabel("continent")
plt.ylabel("wine_servings")

# Third subplot
plt.subplot(2, 2, 3)
plt.scatter(df["spirit_servings"], df["wine_servings"], c="g")
plt.xlabel("spirit_servings")
plt.ylabel("wine_servings")

# Fourth subplot
plt.subplot(2, 2, 4)
plt.bar(df["total_litres_of_pure_alcohol"], df["continent"])
plt.xlabel("total_litres_of_pure_alcohol")
plt.ylabel("continent")

# Display the plot
plt.tight_layout()
plt.show()
```
