```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv("Online Sales Data.csv")
df.head()
```

```python
df.value_counts(['Product Category'])
```

```python
df.groupby(['Product Category'])['Total Revenue'].sum()
```

```python
# Calculate category-wise sum
category_sum = df.groupby('Product Category').sum().reset_index()
category_sum
```

```python
#plt.figure(figsize=(8, 4))
plt.figure(figsize=(8, 4), facecolor='#E2DAB8')
M = ('#D52022', '#060270', '#287B07', '#F8CD1D', '#A77805', '#07BC01')
plt.bar(category_sum['Product Category'], category_sum['Total Revenue'], width = 0.6, color = M)
plt.title("Product Category Wise Total Revenue", fontsize=15, color="#AB0808", fontweight='bold')
plt.xlabel("Product Category", fontsize=15, color="#0A0752", fontweight='bold')
plt.ylabel("Revenue", fontsize=15, color="#0A0752", fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
```
