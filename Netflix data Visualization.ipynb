```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
```

```python
caliHouse = fetch_california_housing()
```

```python
caliHouse
```

```python
data = pd.DataFrame(data=np.c_[caliHouse['data'], caliHouse['target']], columns=caliHouse['feature_names']+['target'])
data
```

```python
X3 = data.drop("target", axis=1)
Y3 = data["target"]
```

```python
x3_train, x3_test, y3_train, y3_test = train_test_split(X3, Y3, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()
model.fit(x3_train, y3_train)
```

```python
y3_pred = model.predict(x3_test)
y3_pred
```

```python
mae3 = mean_absolute_error(y3_test, y3_pred)
mse3 = mean_squared_error(y3_test, y3_pred)
r23 = r2_score(y3_test, y3_pred)

print(mae3)
print(mse3)
print(r23)
```
